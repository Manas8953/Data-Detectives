```python
# Initialize an empty dictionary to store the results for each team
team_stats = {}

# List of teams to analyze
teams = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan',
         'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']

# Loop through each team to calculate the required stats
for team in teams:
    # Filter matches where the team played
    team_matches = df[(df['team1'] == team) | (df['team2'] == team)]
    played = len(team_matches['match_number'].unique())
    
    # Calculate wins and losses
    wins = team_matches[team_matches['winner'] == team]['match_number'].nunique()
    losses = played - wins

    # Filter deliveries for the team's matches
    team_deliveries = df[df['match_number'].isin(team_matches['match_number'])]
    
    # Calculate runs scored and overs faced by the team
    runs_scored = team_deliveries[team_deliveries['Deliveries.batting_team'] == team]['Deliveries.runs_off_bat'].sum()
    balls_faced = team_deliveries[(team_deliveries['Deliveries.batting_team'] == team) & (team_deliveries['Deliveries.wides'] == 0)]['Deliveries.ball'].count()
    overs_faced = balls_faced / 6 if balls_faced > 0 else 0
    
    # Calculate runs conceded and overs bowled by the team
    runs_conceded = team_deliveries[team_deliveries['Deliveries.bowling_team'] == team]['Deliveries.runs_off_bat'].sum()
    balls_bowled = team_deliveries[(team_deliveries['Deliveries.bowling_team'] == team) & (team_deliveries['Deliveries.wides'] == 0)]['Deliveries.ball'].count()
    overs_bowled = balls_bowled / 6 if balls_bowled > 0 else 0
    
    # Calculate NRR, NRR For, and NRR Against (ensuring no division by zero)
    nrr_for = runs_scored / overs_faced if overs_faced > 0 else 0
    nrr_against = runs_conceded / overs_bowled if overs_bowled > 0 else 0
    nrr = nrr_for - nrr_against
    
    # Calculate Extras and % of Total Runs Given in Extras
    extras = team_deliveries[team_deliveries['Deliveries.bowling_team'] == team][['Deliveries.wides', 'Deliveries.noballs', 'Deliveries.byes', 'Deliveries.legbyes', 'Deliveries.penalty']].sum().sum()
    total_runs_conceded = team_deliveries[team_deliveries['Deliveries.bowling_team'] == team]['Deliveries.runs_off_bat'].sum()
    extras_percentage = (extras / total_runs_conceded * 100) if total_runs_conceded > 0 else 0
    
    # Store the results in the dictionary
    team_stats[team] = {
        'Played': played,
        'Won': wins,
        'Lost': losses,
        'NRR': round(nrr, 2),
        'NRR For': round(nrr_for, 2),
        'NRR Against': round(nrr_against, 2),
        'Extras': extras,
        '% Total Runs Given in Extras': round(extras_percentage, 2)
    }

# Convert the dictionary to a DataFrame for easy viewing
results_df = pd.DataFrame(team_stats).T.reset_index().rename(columns={'index': 'Team'})
print(results_df)
```

```python
# Constants for England and New Zealand from the provided table
england_runs_scored = 2795
england_overs_faced = 475
england_runs_given = 2768
england_overs_bowled = 533
england_current_nrr = 0.690965

# New Zealand's NRR target and hypothetical score in the final inning
nz_nrr = 1.157736
nz_score_in_final_inning = 316

# Define function to calculate the required overs for England to qualify
def calculate_required_overs_for_qualification(
    target_runs, runs_scored, overs_faced, runs_given, overs_bowled, target_nrr):
    """
    Calculate the minimum overs required to achieve a target NRR.
    
    Parameters:
    - target_runs: The target score to chase.
    - runs_scored: Total runs scored by the team so far.
    - overs_faced: Total overs faced by the team so far.
    - runs_given: Total runs conceded by the team so far.
    - overs_bowled: Total overs bowled by the team so far.
    - target_nrr: The NRR that the team needs to achieve to qualify.
    
    Returns:
    - required_overs: Minimum overs within which the team must chase the target to achieve the required NRR.
    """
    # Loop to find the minimum overs required, starting from 50 and going down
    for overs in range(50, 0, -1):
        # Calculate the projected NRR if England chases the target in `overs` overs
        projected_nrr = ((runs_scored + target_runs) / (overs_faced + overs)) - (runs_given / overs_bowled)
        if projected_nrr > target_nrr:
            return overs
    return None  # If no solution is found within 50 overs

# Calculate the required overs for England to qualify
required_overs_to_qualify = calculate_required_overs_for_qualification(
    target_runs=316,
    runs_scored=england_runs_scored,
    overs_faced=england_overs_faced,
    runs_given=england_runs_given,
    overs_bowled=england_overs_bowled,
    target_nrr=nz_nrr
)

if required_overs_to_qualify:
    print(f"England needs to chase 316 runs in {required_overs_to_qualify} overs or less to qualify.")
else:
    print("It's not possible for England to qualify based on NRR within 50 overs.")
```
